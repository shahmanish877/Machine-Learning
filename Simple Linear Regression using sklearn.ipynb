```python
# Import required libraries
import numpy as np
np.random.seed(11)

# Prepare toy dataset using equation y = 2x1 + 5
x1 = np.random.randn(100)
y = 2 * x1 + 5 + np.random.randn(100) * 0.9
```

```python
# Plot the graph
import matplotlib.pyplot as plt
plt.scatter(x1, y)
plt.show()
```

```python
X = np.array(x1).reshape((-1, 1))
print(X.shape)
```

```python
# Split given data into training data and test data
# Use sklearn.model_selection import train_test_split
from sklearn.model_selection import train_test_split


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.2, random_state=11)

print(X_train.shape, y_train.shape)
# Train model in train data
# Test the prediciton of model in test data
```

```python
from sklearn.linear_model import LinearRegression

# Create Linear regressor model
linear_model = LinearRegression().fit(X_train, y_train)
```

```python
# Predict on training data
y_pred = linear_model.predict(X_test)

# Plot the predicted outcome
plt.scatter(X_test, y_test, marker = '.')
plt.plot(X_test, y_pred, color = 'orange')
plt.show()
```

```python
print("R2 :", linear_model.score(X, y))
```

```python
print("Regression model coefficients are: " + str(linear_model.coef_[0]) + " and " + str(linear_model.intercept_))
```

```python
from sklearn.metrics import mean_squared_error

print("Mean squared error is " + str(mean_squared_error(y, y_pred)))
print("Root Mean Squared Error is " + str(np.sqrt(mean_squared_error(y, y_pred))))
```
